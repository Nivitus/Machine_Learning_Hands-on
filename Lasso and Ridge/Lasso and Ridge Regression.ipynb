```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Collecting the dataset from sklearn
from sklearn.datasets import load_boston
```

```python
# Load the boston dataset
df = load_boston()
```

```python
# Print the dataset
df
```

```python
# Conver to the data frame
dataset = pd.DataFrame(df.data)
print(dataset.head())
```

```python
dataset.columns = df.feature_names
```

```python
dataset.head() 
```

```python
dataset.shape
```

```python
df.target.shape
```

```python
dataset['Price'] = df.target
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
# Linear Regression
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

model = LinearRegression()
mse = cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=5)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

rid = Ridge()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
rid_reg = GridSearchCV(rid,parameters,scoring='neg_mean_squared_error',cv=5)
rid_reg.fit(X,y)
```

```python
print(rid_reg.best_params_)
print(rid_reg.best_score_)
```

```python
# Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

las = Lasso()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
las_reg = GridSearchCV(las,parameters,scoring='neg_mean_squared_error',cv=5)
las_reg.fit(X,y)
```

```python
print(las_reg.best_params_)
print(las_reg.best_score_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
pred_lasso = las_reg.predict(X_test)
pred_ridg = rid_reg.predict(X_test)
```
