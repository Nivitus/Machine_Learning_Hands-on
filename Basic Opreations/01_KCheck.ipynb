Q9) Create a function age_group which does the following:


     Less than 18 =1
     18-35        =2
     36-50        =3
     51+          =4    
     
     
     Create a new variable "age_cat" and use the above age_group function in df1['age'] column 
    
    

```python
def age_cat(x):
    if x <= 18:
        return 1
    elif x >= 19 and x <=35:
        return 2
    elif x >= 36 and x <=50:
        return 3
    elif x >= 51:
        return 5
    else:
        return x
```

```python
df1['age_cat'] = df1.age.apply(age_cat)
df1[['age','age_cat']]
```
